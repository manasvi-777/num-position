```python
'''Get a Python list of N integer values.
Find the presence of ODD and EVEN numbers in the list.
If the ODD number occurs at the ODD position replace the ODD number with 'odd'.
If the EVEN number occurs at the EVEN position replace the EVEN number with 'even'.

Find the number of odd places that were changed and also the number of even places that were changed

Prepare a new list for the odd and even positions.


Sample Input:
5 
231
453
806
1240
3228

Output: 

['odd', 453, 806, 'even', 3228]

odd=1

oddpositionlist=[0]

even=1

evenpositionlist=[3]

Test Case1:

Input:

5
23
45
80
12
32

Output:

['odd', 45, 80, 'even', 32]
odd= 1
oddpositionlist= [0]
even= 1
evenpositionlist= [3]

Test Case2:

Input:

6
23
44
3
12
71
33

Output:

['odd', 'even', 'odd', 'even', 'odd', 33]
odd= 3
oddpositionlist= [0, 2, 4]
even= 2
evenpositionlist= [1, 3]
'''
```

```python
N=int(input())
l=[]
even=0
odd=0
even_position_list=[]
odd_position_list=[]
for i in range(N):
    num=int(input())
    l.append(num)
    
for i in range(len(l)):
    position=i+1
    val=str(l[i])
    if position%2==0:
        position="even"
        if int(val[-1])%2==0:
            l[i]=position
            even+=1
            even_position_list+=[i]
    else:
        position="odd"
        if int(val[-1])%2!=0:
            l[i]=position
            odd+=1
            odd_position_list+=[i]
        
print(l)
print("odd= ",odd)
print("oddpositionlist= ",odd_position_list)
print("even=",even)
print("evenpositionlist= ",even_position_list)
```
